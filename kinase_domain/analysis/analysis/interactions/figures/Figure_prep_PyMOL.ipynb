Load the receptor and ligand

```python
PyMOL>load docking-5LJJ/data/receptor/receptor_clean.pdb, receptor
PyMOL>load docking-5LJJ/data/ligands/native_ligand.pdb, ligand
```

Prepare the molecule

```python
PyMOL>show sticks, ligand
PyMOL>select hbonds_residues, receptor and resi 603+605
PyMOL>show sticks, hbonds_residues
PyMOL>distance hbonds, ligand, hbonds_residues, 3.5
```

Clean the receptor and ligand to make it figure-ready

```python
PyMOL>hide labels
PyMOL>bg_color white
PyMOL>set ray_trace_mode, 1
PyMOL>set antialias, 3
PyMOL>set hash_max, 300
PyMOL>set dash_gap, 0.3
PyMOL>set dash_length, 0.5
PyMOL>set dash_radius, 0.05
PyMOL>ray 
PyMOL>png 5LJJ-{ligand}.png, dpi=600
```
